```python
import numpy as np #liner algebra
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/Iris.csv") # df= data frame
df.head() # first 5
```

```python
df.tail() #last 5
```

```python
#Rename Columns
df.rename({'SepalLengthCm': 'sepal_length',
           'SepalWidthCm' : 'sepal_width',
           'PetalLengthCm': 'petal_length',
           'PetalWidthCm' : 'petal_width',
           'Species' : 'species'
           }, inplace=True,axis=1) #axis is column
df.sample(5)
```

```python
x= df.groupby('species').median() #median is out layer
x
```

```python
#statistics of this dataset
stats = df.describe()
stats
```

**Aggregation Function**

```python
df.groupby('species').agg(['mean', 'median']) #passing a list of dataset
df.groupby('species').agg([np.mean, np.median])
```

**Visualisations**

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Distribution Plot

```python
sns.distplot(a=df['petal_width'],bins=40,color='y')
plt.title('Distribution of petal-width')
```

```python
sns.countplot(x='species',data=df,palette='husl') #Count plot

```

Box plot


```python
plt.figure(figsize=(10,7))
sns.boxplot(x='species' ,y='sepal_width',data=df ,palette='YlGnBu')
```

```python
sns.heatmap(df.corr(),cmap="rocket", linecolor="white", linewidth=1)
```

```python
x=df.drop('species',axis=1)
g=sns.clustermap(x ,cmap="magma")
```

```python
sns.regplot(x='petal_width', y='petal_length', data=df) #liner regression
```

Scatter plot


```python
ax = plt.axes()

ax.scatter(df.sepal_length, df.sepal_width)

#label the axes
ax.set(xlabel='Sepal Length (cm)',
       ylabel='Sepal Width (cm)',
       title='Sepal Length vs Width');
```

```python
ax = plt.axes()

ax.scatter(df.petal_length, df.petal_width)

#label the axes
ax.set(xlabel='petal Length (cm)',
       ylabel='petal Width (cm)',
       title='petal Length vs Width');
```
